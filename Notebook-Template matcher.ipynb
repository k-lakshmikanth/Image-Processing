```python
# Read the images as gray scale
img1 = cv2.imread("magazine.JPG",cv2.IMREAD_GRAYSCALE)
img2= cv2.imread("magazine_with_distractors2.JPG",cv2.IMREAD_GRAYSCALE)
```

```python
# To view the images uncomment the below lines
# plt_img(img1)
# plt_img(img2)
```

```python
# Viewing the images
cv2.imshow("img1",img1)
cv2.imshow("img2",img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Initiate the ORB detector
orb = cv2.ORB_create()

# Find the keypoints and descriptors with ORB
kp1, des1 = orb.detectAndCompute(img1,None)
kp2, des2 = orb.detectAndCompute(img2,None)
```

```python
# Initiate the BFMatcher
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)

# Match descriptors
matches = bf.match(des1,des2)

# Sort them in the order of their distance
matches = sorted(matches, key = lambda x:x.distance)
```

```python
# Draw first 20 matches
img3 = cv2.drawMatches(img1,kp1,img2,kp2,matches[:30],None,flags=2)

plt_img(img3,(25,25))
```
